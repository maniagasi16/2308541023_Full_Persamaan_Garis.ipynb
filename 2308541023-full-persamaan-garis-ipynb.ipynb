```python
import numpy as np
import matplotlib.pyplot as plt

# Meminta pengguna memilih kasus
kasus = int(input("Kasus berapakah yang akan dikerjakan? (1, 2, atau 3) "))

if kasus == 1:
    print("Masing-masing garis akan melewati dua titik yang diketahui.")
    # Garis pertama
    a = 2
    b = 3
    c = 4
    d = 5
    # Garis kedua
    e = 1
    f = 2
    g = 3
    h = 4

    m_1 = (d - b) / (c - a)  # Gradien garis pertama
    m_2 = (h - f) / (g - e)  # Gradien garis kedua

    # Persamaan garis pertama adalah y = m_1 * x + c_1
    c_1 = b - m_1 * a  # Titik potong garis pertama terhadap sumbu y
    # Persamaan garis kedua adalah y = m_2 * x + c_2
    c_2 = f - m_2 * e  # Titik potong garis kedua terhadap sumbu y

    print("Persamaan garis pertama adalah y = {:.3f}x + {:.3f}".format(m_1, c_1))
    print("Persamaan garis kedua adalah y = {:.3f}x + {:.3f}".format(m_2, c_2))

    if m_1 == m_2:
        if c_1 == c_2:
            print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi.")
        else:
            print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
    else:
        tp_x = (c_2 - c_1) / (m_1 - m_2)
        tp_y = m_1 * tp_x + c_1
        print("Titik potong kedua garis adalah ({:.3f}, {:.3f})".format(tp_x, tp_y))

    x = np.linspace(min(a, c, e, g) - 5, max(a, c, e, g) + 5, 100)
    plt.plot(x, m_1 * x + c_1, color='red', label="y = {:.3f}x + {:.3f}".format(m_1, c_1))
    plt.plot(x, m_2 * x + c_2, color='blue', label="y = {:.3f}x + {:.3f}".format(m_2, c_2))
    if m_1 != m_2:
        plt.scatter(tp_x, tp_y, color='green', label="Titik potong ({:.3f}, {:.3f})".format(tp_x, tp_y))
    plt.axvline(x=0, color='grey', linewidth=0.5)
    plt.axhline(y=0, color='grey', linewidth=0.5)
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

elif kasus == 2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    # Garis pertama
    m_1 = 2
    a = 1
    b = 4
    # Garis kedua
    m_2 = -1
    c = 2
    d = 3

    c_1 = b - m_1 * a
    c_2 = d - m_2 * c

    print("Persamaan garis pertama adalah y = {:.3f}x + {:.3f}".format(m_1, c_1))
    print("Persamaan garis kedua adalah y = {:.3f}x + {:.3f}".format(m_2, c_2))

    if m_1 == m_2:
        if c_1 == c_2:
            print("Terdapat tak berhingga banyaknya solusi.")
        else:
            print("Tidak terdapat solusi.")
    else:
        tp_x = (c_2 - c_1) / (m_1 - m_2)
        tp_y = m_1 * tp_x + c_1
        print("Titik potong kedua garis adalah ({:.3f}, {:.3f})".format(tp_x, tp_y))

    x = np.linspace(min(a, c) - 5, max(a, c) + 5, 100)
    plt.plot(x, m_1 * x + c_1, color='red', label="y = {:.3f}x + {:.3f}".format(m_1, c_1))
    plt.plot(x, m_2 * x + c_2, color='blue', label="y = {:.3f}x + {:.3f}".format(m_2, c_2))
    if m_1 != m_2:
        plt.scatter(tp_x, tp_y, color='green', label="Titik potong ({:.3f}, {:.3f})".format(tp_x, tp_y))
    plt.axvline(x=0, color='grey', linewidth=0.5)
    plt.axhline(y=0, color='grey', linewidth=0.5)
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

elif kasus == 3:
    print("Kedua garis memenuhi persamaan berbentuk ax + by = c dan dx + ey = f.")
    # Garis pertama
    a = 1
    b = -2
    c = 1
    # Garis kedua
    d = 2
    e = 3
    f = 12

    # Menentukan determinan
    det = a * e - b * d

    if det == 0:
        if a * f == c * d:
            print("Kedua garis berhimpitan, terdapat tak berhingga banyaknya solusi.")
        else:
            print("Kedua garis sejajar, tidak terdapat solusi.")
    else:
        tp_x = (c * e - b * f) / det
        tp_y = (a * f - c * d) / det
        print("Titik potong kedua garis adalah ({:.3f}, {:.3f})".format(tp_x, tp_y))

    x = np.linspace(tp_x - 5, tp_x + 5, 100)
    y1 = (c - a * x) / b
    y2 = (f - d * x) / e
    plt.plot(x, y1, color='red', label="{:.3f}x + {:.3f}y = {:.3f}".format(a, b, c))
    plt.plot(x, y2, color='blue', label="{:.3f}x + {:.3f}y = {:.3f}".format(d, e, f))
    if det != 0:
        plt.scatter(tp_x, tp_y, color='green', label="Titik potong ({:.3f}, {:.3f})".format(tp_x, tp_y))
    plt.axvline(x=0, color='grey', linewidth=0.5)
    plt.axhline(y=0, color='grey', linewidth=0.5)
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

else:
    print("Hanya terdapat 3 kasus. Mohon masukkan 1, 2 atau 3 saja.")
```
